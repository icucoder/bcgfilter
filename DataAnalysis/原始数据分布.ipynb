```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import torch
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns

Pathlist = [
    '/root/zqh/NewDataSet/New_data_pa8.pt',
    '/root/zqh/NewDataSet/New_data_pa9.pt',
    '/root/zqh/NewDataSet/New_data_pa3.pt',
    '/root/zqh/NewDataSet/New_data_pa5.pt',
    '/root/zqh/NewDataSet/New_data_pa6.pt',
    '/root/zqh/NewDataSet/New_data_pa7.pt',
]


for i in range(len(Pathlist)):
    data = torch.load(Pathlist[i])
    data = data.squeeze(1) # 30 300
    pca = PCA(n_components=1)
    data = pca.fit_transform(data) # shape: 30,1
    var = np.var(data)
    data = data/(var**0.5)
    plt.subplot(3,2,i+1)
    plt.xlim(-3, 3)
    plt.ylim(0, 1)
    sns.kdeplot(data[:,0:1])
plt.show()
```
